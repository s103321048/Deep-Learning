```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
# 3D
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator
from collections import OrderedDict
```

### Example

$
f(x, y) = \frac{1}{20} x ^ 2 + y ^ 2
$

```python
def f(x, y):
    return (0.05 * x ** 2) + (y ** 2)

def df(x, y):
    return (0.1 * x) , (2.0 * y)

x0 = np.arange(-10, 11, 1)
x1 = np.arange(-10, 11, 1)
X0, X1 = np.meshgrid(x0, x1)
Y = f(X0, X1)

fig = plt.figure()
ax = fig.gca(projection = '3d')

surf = ax.plot_surface(X0, X1, Y)
plt.show()
```

# SGD

$
W \leftarrow W - \eta\frac{\partial L}{\partial W}
$

```python
class SGD:
    def __init__(self, lr = 0.01):
        self.lr = lr
        
    def update(self, params, grads):
        for key in params.keys():
            params[key] -= self.lr * grads[key] 
```

```python
optimizer = SGD(lr = 0.95)
idx = 1
init_pos = (-10.0, 2.5)
params = {}
params['x'], params['y'] = init_pos[0], init_pos[1]
grads = {}
grads['x'], grads['y'] = 0, 0

x_history = []
y_history = []
for i in range(100):
    x_history.append(params['x'])
    y_history.append(params['y'])
    grads['x'], grads['y'] = df(params['x'], params['y'])
    optimizer.update(params, grads)

x = np.arange(-11, 11, 0.01)
y = np.arange(-5, 5, 0.01)

X, Y = np.meshgrid(x, y) 
Z = f(X, Y)

mask = Z > 7
Z[mask] = 0
idx += 1
plt.plot(x_history, y_history, 'o-', color="red")
plt.contour(X, Y, Z)
plt.ylim(-11, 11)
plt.xlim(-11, 11)
plt.plot(0, 0, '+')
plt.title(optimizer.__class__.__name__)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# Momentum

$
xv \leftarrow \alpha - \eta\frac{\partial L}{\partial W}
$

$
W \leftarrow W + v
$

```python
class Momentum:
    def __init__(self, lr = 0.01, momentum = 0.9):
        self.lr = lr
        self.momentum = momentum
        self.v = None
        
    def update(self, params, grads):
        if self.v is None:
            self.v = {}
            for key, val in params.items():
                self.v[key] = np.zeros_like(val)
        for key in params.keys():
            self.v[key] = (self.momentum * self.v[key]) - (self.lr * grads[key])
            params[key] += self.v[key]
```

```python
optimizer = Momentum(lr = 0.1)
idx = 1
init_pos = (-10.0, 2.5)
params = {}
params['x'], params['y'] = init_pos[0], init_pos[1]
grads = {}
grads['x'], grads['y'] = 0, 0

x_history = []
y_history = []
for i in range(100):
    x_history.append(params['x'])
    y_history.append(params['y'])
    grads['x'], grads['y'] = df(params['x'], params['y'])
    optimizer.update(params, grads)

x = np.arange(-11, 11, 0.01)
y = np.arange(-5, 5, 0.01)

X, Y = np.meshgrid(x, y) 
Z = f(X, Y)

mask = Z > 7
Z[mask] = 0
idx += 1
plt.plot(x_history, y_history, 'o-', color="red")
plt.contour(X, Y, Z)
plt.ylim(-11, 11)
plt.xlim(-11, 11)
plt.plot(0, 0, '+')
plt.title(optimizer.__class__.__name__)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# AdaGrad

$
h \leftarrow h + \eta\frac{\partial L}{\partial W} \odot \eta\frac{\partial L}{\partial W}
$

$
W \leftarrow W - \eta\frac{1}{\sqrt{h}} \frac{\partial L}{\partial W}
$

```python
class AdaGrad:
    def __init__(self, lr = 0.01):
        self.lr = lr
        self. h = None
    
    def update(self, params, grads):
        if self.h is None:
            self.h = {}
            for key, val in params.items():
                self.h[key] = np.zeros_like(val)
        for key in params.keys():
            self.h[key] += grads[key] * grads[key]
            params[key] -= (self.lr * grads[key]) / (np.sqrt(self.h[key]) + 1e-7)
```

```python
optimizer = AdaGrad(lr = 1.5)
idx = 1
init_pos = (-10.0, 2.5)
params = {}
params['x'], params['y'] = init_pos[0], init_pos[1]
grads = {}
grads['x'], grads['y'] = 0, 0

x_history = []
y_history = []
for i in range(100):
    x_history.append(params['x'])
    y_history.append(params['y'])
    grads['x'], grads['y'] = df(params['x'], params['y'])
    optimizer.update(params, grads)

x = np.arange(-11, 11, 0.01)
y = np.arange(-5, 5, 0.01)

X, Y = np.meshgrid(x, y) 
Z = f(X, Y)

mask = Z > 7
Z[mask] = 0
idx += 1
plt.plot(x_history, y_history, 'o-', color="red")
plt.contour(X, Y, Z)
plt.ylim(-11, 11)
plt.xlim(-11, 11)
plt.plot(0, 0, '+')
plt.title(optimizer.__class__.__name__)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# Adam

# http://arxiv.org/abs/1412.6980v8

```python
class Adam:
    def __init__(self, lr=0.001, beta1=0.9, beta2=0.999):
        self.lr = lr
        self.beta1 = beta1
        self.beta2 = beta2
        self.iter = 0
        self.m = None
        self.v = None
        
    def update(self, params, grads):
        if self.m is None:
            self.m, self.v = {}, {}
            for key, val in params.items():
                self.m[key] = np.zeros_like(val)
                self.v[key] = np.zeros_like(val)        
        self.iter += 1
        lr_t  = self.lr * np.sqrt(1.0 - (self.beta2 ** self.iter)) / (1.0 - (self.beta1 ** self.iter))                 
        for key in params.keys():
            self.m[key] += (1 - self.beta1) * (grads[key] - self.m[key])
            self.v[key] += (1 - self.beta2) * ((grads[key] ** 2) - self.v[key])            
            params[key] -= lr_t * self.m[key] / (np.sqrt(self.v[key]) + 1e-7)            
```

```python
optimizer = Adam(lr = 0.3)
idx = 1
init_pos = (-10.0, 2.5)
params = {}
params['x'], params['y'] = init_pos[0], init_pos[1]
grads = {}
grads['x'], grads['y'] = 0, 0

x_history = []
y_history = []
for i in range(100):
    x_history.append(params['x'])
    y_history.append(params['y'])
    grads['x'], grads['y'] = df(params['x'], params['y'])
    optimizer.update(params, grads)

x = np.arange(-11, 11, 0.01)
y = np.arange(-5, 5, 0.01)

X, Y = np.meshgrid(x, y) 
Z = f(X, Y)

mask = Z > 7
Z[mask] = 0
idx += 1
plt.plot(x_history, y_history, 'o-', color="red")
plt.contour(X, Y, Z)
plt.ylim(-11, 11)
plt.xlim(-11, 11)
plt.plot(0, 0, '+')
plt.title(optimizer.__class__.__name__)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
