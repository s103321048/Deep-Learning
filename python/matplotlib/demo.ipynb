```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

# 制圖

```python
x = np.arange(0, 10, 0.1)
print(x.shape)
x
```

```python
y = np.sin(x)
print(y.shape)
y
```

```python
z = np.cos(x)
z
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, z)
plt.show()
```

```python
plt.plot(x, y, label = 'Sin')
plt.plot(x, z, label = 'Cos', linestyle = '--')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sin & Cos')
plt.legend() #圖例
plt.show()
```
